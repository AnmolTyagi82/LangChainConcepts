```python
from langchain.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import OpenAIEmbeddings
from langchain.llms import OpenAI
from langchain.vectorstores import Pinecone as PC
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
import os
```

```python
loader = PyPDFDirectoryLoader("pdfs")
```

```python
data = loader.load()
```

```python
data
```

```python
data[0]
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
```

```python
text_chunks = text_splitter.split_documents(data)
```

```python
OPENAI_API_KEY = os.environ["OPENAI_API_KEY"]
```

```python
embeddings = OpenAIEmbeddings()
```

```python
embeddings.embed_query("How are you?")
```

```python
len(embeddings.embed_query("How are you?"))
```

```python
PINECONE_API_KEY = os.environ["PINECONE_API_KEY"]
PINECONE_API_ENV = os.environ["PINECONE_API_ENV"]
```

```python
from pinecone import Pinecone, ServerlessSpec
```

```python
pc = Pinecone(api_key=PINECONE_API_KEY)
```

```python
index_name = "testing"
```

```python
index = pc.Index(index_name)
```

```python
docsearch = PC.from_texts([t.page_content for t in text_chunks], embeddings, index_name=index_name)
```

```python
docsearch
```

```python
query = "YOLOv7 outperforms which model"
```

```python
docs = docsearch.similarity_search(query)
```

```python
docs
```

```python
llm = OpenAI()
```

```python
docsearch.as_retriever()
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=docsearch.as_retriever())
```

```python
query = "YOLOv7 outperforms which model?"
```

```python
qa.invoke(query)
```

```python
import sys
while True:
    user_input = input(f"Input Prompt: ")
    if user_input == "exit":
        print("Existing")
        sys.exit()
    if user_input == "":
        continue
    result = qa.invoke({"query": user_input})
    print(f"Answer: {result['result']}")
```
